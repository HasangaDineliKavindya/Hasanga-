```python
import pandas as pd
from simpletransformers.seq2seq import Seq2SeqModel, Seq2SeqArgs
```

```python
model_args = Seq2SeqArgs()
model_args.num_train_epochs = 10
model_args.no_save = True
model_args.evaluate_generated_text = True
model_args.evaluate_during_training = True
model_args.evaluate_during_training_verbose = True

# Initialize model without CUDA
model = Seq2SeqModel(
    encoder_decoder_type="bart",
    encoder_decoder_name="facebook/bart-large",
    args=model_args,
    use_cuda=False,
)

```

```python
import pandas as pd

train_data = [
    [
        "Perseus 'Percy' Jackson is the main protagonist and the narrator of the Percy Jackson and the Olympians series.",
        "Percy is the protagonist of the Percy Jackson and the Olympians.",
    ],
    [
        "Annabeth Chase is one of the main protagonists in Percy Jackson and the Olympians.",
        "Annabeth is a protagonist in the Percy Jackson and the Olympians.",
    ],
]

train_df = pd.DataFrame(
    train_data, columns=["input_text", "target_text"]
)
```

```python
eval_data = [
    [
        "Grover Underwood is a satyr and the Lord of the Wild. He is the satyr who found the demigods Thalia Grace, Nico and Bianca di Angelo, Percy Jackson, Annabeth Chase, and Luke.",
        "Grover is a satyr who found many important demigods.",
    ],
    [
       "Thalia Grace is the daughter of Zeus, sister of Jackson Grace. After several years as a pine tree on Half-Blood Hill, she got a new job leading the Hunters of Artemis." ,
       "Thalia is the daughter of Zeus and leader of the Hunters of Artemis.",
    ],
]

eval_df = pd.DataFrame(eval_data, columns=["input_text", "target_text"])
```

```python
train_df
```

```python
eval_df
```

```python
def count_matches(labels,preds):
  print(labels)
  print(preds)
  return sum([
      1 if label ==pred else 0
      for label,pred in zip (labels,preds)
  ])
```

```python
#train model
model.train_model(train_df,eval_data=eval_df,matches=count_matches)
```

```python
# Evaluate the model
result = model.eval_model(eval_df)
```

```python
result
```

```python
#use the model for predicttion

print(model.predict(
    [
        "Tyson is a Cyclops,a son of Poseidon,and Percy Jackson's half brother.He is the current geanaral of the Cyclopes army."
    ]
))
```
